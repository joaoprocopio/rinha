```python

import json
import os
from pathlib import Path
import pandas as pd

exec_dir = Path(os.getcwd())
root_dir = exec_dir.parent

data_path = root_dir.joinpath("./ext/previa-resultados+participantes-info.json")

with open(data_path) as raw_data:
    data = json.load(raw_data)
```

```python
df = pd.json_normalize(data)

pd.set_option('display.max_columns', None)  # Mostra todas as colunas
pd.set_option('display.max_rows', None)     # Mostra todas as linhas
pd.set_option('display.width', None)        # Largura máxima do display
pd.set_option('display.max_colwidth', None) # Mostra conteúdo completo das células
```
